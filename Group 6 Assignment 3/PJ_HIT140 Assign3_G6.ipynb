#### Fit multiple linear regression model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Select relevant features for the model
# Specify the columns to exclude
columns_to_exclude = ['total_updrs', 'motor_updrs']

# Select all variables except the ones to exclude
selected_variables = data1.drop(columns=columns_to_exclude)

X = selected_variables
y_motor = data1['motor_updrs']
y_total = data1['total_updrs']

# Split the data into train and test sets
X_train, X_test, y_motor_train, y_motor_test, y_total_train, y_total_test = train_test_split(
    X, y_motor, y_total, test_size=0.2, random_state=42)

# Create and train the regression models
motor_model = LinearRegression()
total_model = LinearRegression()

motor_model.fit(X_train, y_motor_train)
total_model.fit(X_train, y_total_train)

# Make predictions
y_motor_pred = motor_model.predict(X_test)
y_total_pred = total_model.predict(X_test)

# Evaluate the models
motor_rmse = mean_squared_error(y_motor_test, y_motor_pred, squared=False)
motor_r2 = r2_score(y_motor_test, y_motor_pred)

total_rmse = mean_squared_error(y_total_test, y_total_pred, squared=False)
total_r2 = r2_score(y_total_test, y_total_pred)

print(f'Motor UPDRS Model - RMSE: {motor_rmse}, R-squared: {motor_r2}')
print(f'Total UPDRS Model - RMSE: {total_rmse}, R-squared: {total_r2}')
```
